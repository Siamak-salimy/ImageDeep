```python
import cv2
from google.colab.patches import cv2_imshow
input_image =cv2.imread('1.jpg')
cv2_imshow(input_image)
```

```python
height,width =input_image.shape[0:2]
print('height : ',height)
print(' width :' , width)

height,width,channel =input_image.shape
print('height : ',height)
print(' width :' , width)
print(' channel :' , channel)
print(' size :', input_image.size)
print(' shape :', input_image.shape)
```

```python
''' Resize '''
resized_image = cv2.resize(input_image,(0,0) , fx=2 , fy = 2)
#resized_image = cv2.resize(input_image,(500,500))

cv2_imshow(resized_image)
```

```python
height,width,channel =resized_image.shape
print('height : ',height)
print(' width :' , width)
```

```python
image_2=cv2.cvtColor(resized_image,cv2.COLOR_BGR2GRAY)
image_3=255 - image_2
cv2_imshow(image_2)
cv2_imshow(image_2)
```

```python
image_2=cv2.rotate(resized_image,cv2.ROTATE_90_CLOCKWISE)
cv2_imshow(image_2)
```

```python
flipvertical = cv2.flip(resized_image,0)
cv2_imshow(flipvertical)
```

```python
final=cv2.hconcat((resized_image,flipvertical))
cv2_imshow(final)
```

```python
pixel=resized_image[300,300]
print(pixel)
```

```python
final=cv2.hconcat((image_2,image_3))
cv2_imshow(final)
```
